```python
import pandas as pd
import numpy as np
```

Read Csv Files

```python
df_flights = pd.read_csv ('~/Downloads/C1DC/RAW_Flights.csv')
df_tickets = pd.read_csv ('~/Downloads/C1DC/RAW_Tickets.csv')
df_ac = pd.read_csv ('~/Downloads/C1DC/RAW_AC.csv')
```

Cleaning Missing Values

See how many null values are in each columns

```python
df_flights.isnull().sum()
```

```python
df_tickets.isnull().sum()
```

```python
df_ac.isnull().sum()
```

Drop rows with null values

```python
df_na_flights=df_flights.dropna(how='any')
df_na_tickets=df_tickets.dropna(how='any')
df_na_ac=df_ac.dropna(how='any')
```

See how many duplicates are in each data frame

```python
df_na_tickets.duplicated().sum()
```

```python
df_na_ac.duplicated().sum()
```

```python
df_na_ac[df_na_ac['IATA_CODE'].duplicated()].head()
```

Drop Duplicates in the data sets

```python
df_dd_flights=df_na_flights.drop_duplicates()
df_dd_tickets=df_na_tickets.drop_duplicates()
df_dd_ac=df_na_ac.drop_duplicates() 
```

Export Clean CSV files

```python
df_dd_flights.to_csv('~/Downloads/C1DC/Clean_Flights.csv', index=False)
df_dd_tickets.to_csv('~/Downloads/C1DC/Clean_Tickets.csv',index=False)
df_dd_ac.to_csv('~/Downloads/C1DC/Clean_AC.csv',index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read Clean CSV files

```python
df_clean_flights = pd.read_csv ('~/Downloads/C1DC/Clean_Flights.csv')
df_clean_tickets = pd.read_csv ('~/Downloads/C1DC/Clean_Tickets.csv')
df_clean_ac = pd.read_csv ('~/Downloads/C1DC/Clean_AC.csv')
```

10 Busiest Round Trip Routes in 1Q2019

```python
df_clean_flights.head()
```

Join Origin and Destination Columns together to create a new column OD (Origin and Destination)

```python
df_clean_flights['OD'] = df_clean_flights[['ORIGIN','DESTINATION']].apply(lambda x:' '.join(sorted(x)), axis=1)
```

```python
df_clean_flights
```

Removing Flights that are Canceled

```python
df_nacancel_flights = df_clean_flights[df_clean_flights['CANCELLED'] == 0.0]
```

```python
df_nacancel_flights
```

Group by Origin and Destination Column together

```python
df_grouped_flights = df_nacancel_flights.groupby('OD', as_index=False)
```

Count Each column by OD (Origin and Destination)

```python
df_count_flights = df_grouped_flights.count()
```

Count the top 10 busiest round trip routes

```python
df_largest_flights = df_count_flights.nlargest(10,'OP_CARRIER')
```

```python
df_largest_flights
```

This Bar graph shows the 10 Busiest round trips in terms of number of round trips in 1Q2019

```python
df_largest_flights.plot.bar(x='OD', y='OP_CARRIER')
```
